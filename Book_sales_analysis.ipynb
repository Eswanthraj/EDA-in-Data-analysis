```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year')
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("No of Books")
plt.title("No of Books in each Genre")
plt.show()
```

```python
df.groupby('Author')['Book_average_rating'].mean().sort_values(ascending=False).head(10).plot(kind='barh')
plt.xlabel('Author')
plt.ylabel('Average Rating')
plt.title('Top 10 Authors by Average Rating')
plt.show()
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending = False).max()
```
